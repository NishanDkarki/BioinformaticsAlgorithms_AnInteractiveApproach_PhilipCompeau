```python
def LCSBackTrack(v, w):
    n = len(v)
    m = len(w)
    s = np.array([0]*((n+1)*(m+1))).reshape(n+1,m+1)
    Backtrack = np.array([0]*((n+1)*(m+1))).reshape(n+1,m+1)
    
#     for i in range (1, n+1):
#         s[i][0] = 0
#     for j in range (1, m+1):
#         s[0][j] = 0
        
    for i in range (1, n+1):
        for j in range (1, m+1):
            if v[i-1] == w[j-1]:
                s[i][j] = max(s[i-1][j] , s[i][j-1] , s[i-1][j-1] + 1)
            else:
                s[i][j] =max(s[i-1][j], s[i][j-1])
                
            if s[i][j] == s[i-1][j]:
                Backtrack[i][j] = "1" #down
                
            elif s[i][j] == s[i][j-1]:
                Backtrack[i][j] = "2" #right
                
            elif s[i][j] == s[i-1][j-1] + 1:
                Backtrack[i][j] = "3" #diag
    return Backtrack

lcs = []
def OutputLCS(backtrack, v, i, j):

    if i == 0 or j == 0:
        return
    if backtrack[i][j] == 1:
        OutputLCS(backtrack, v, i - 1, j)
    elif backtrack [i][j] == 2:
        OutputLCS(backtrack, v, i, j - 1)
    else:
        OutputLCS(backtrack, v, i - 1, j - 1)
        lcs.append(v[i-1])
        
#     print(lcs)

    return lcs


def main():
    
    with open('/Users/nises/Downloads/dataset_604482_5.txt') as input_data:
        v = input_data.readline().strip()
        w = input_data.readline().strip()
    
#     print(v)
    backtrack = LCSBackTrack(v,w)
#     print(backtrack)
    answer = OutputLCS(backtrack, v, len(v), len(w))
    print ("".join(answer))
    
if __name__ == '__main__':
    main()
```

```python
import networkx as nx

def longest_dag(ed, s_n, e_n):
    d_graph = nx.DiGraph()
    edges = ed
#     edges = [(0,1,{'weight':7}), (0,2,{'weight':4}), (2,3,{'weight':2}), (1,4,{'weight':1}), (3,4,{'weight':3})]
    d_graph.add_edges_from(edges)
#     print(edges)

    paths = nx.all_simple_paths(d_graph, source=s_n, target=e_n)
    # print(list(paths))

    all_paths = []
    for path in paths:
        all_paths = list(set(all_paths + path))
#     print(all_paths)

    for edge in edges:
    #     print(edge)
        if ((edge[0] not in all_paths) | (edge[1] not in all_paths)):
            d_graph.remove_edge(*edge[:2])
#     print(d_graph)

    wt = nx.dag_longest_path(d_graph, weight="weight")
    l_path = nx.dag_longest_path_length(d_graph, weight="weight")

    print(l_path)
#     print(wt)
    return wt

def main():
    with open('/Users/nises/Downloads/dataset_604482_7.txt') as input_data:
        lines = input_data.read().splitlines()
                  
    s_n = int(lines[0])
    e_n = int(lines[1])
    edges = []
    
    for line in lines[2:]:
        a = int(line.split("->")[0])
        b = int(line.split("->")[1].split(':')[0])
        w = int(line.split("->")[1].split(':')[1])
        edge = (a,b,{'weight':w})
        edges.append(edge)

#     print(e_n)
    answer = longest_dag(edges, s_n, e_n)
    answer = map(str, answer)
    print ("->".join(answer))
    
if __name__ == '__main__':
    main()
```
