```python
import os
import collections

def all_combos(k):
    return [''.join(p) for p in itertools.product('ATCG', repeat=k)]

def find_all_kmers(text, k):
    all_kmers = []
    for i in range(len(text)-k+1):
        kmer = text[i:k+i]
        if kmer not in all_kmers:
            all_kmers.append(kmer)
    return all_kmers

def calculate_ham_distance(text1, text2):
    ham_distance = 0
    for i in range(len(text1)):  
        if text1[i] != text2[i]:
            ham_distance += 1
    return ham_distance

def check_seq(dna, kmer, d):
    shared = False
    shares = 0
    
    for seq in dna:
        for i in range(len(seq) - len(kmer)+1):
            test = seq[i:i+len(kmer)]
            if calculate_ham_distance(kmer, test) <= d:
                shares += 1
                break
        
    if shares == len(dna):
        shared = True
    return shared


def motif_gen(text, k, d):
    
    all_motifs = []
    f_kmers = find_all_kmers(text[0], k)
    possible_kmers = all_combos(k)
        
    for pattern in f_kmers:
        pattern1 = []
        for e_kmer in possible_kmers:
            if calculate_ham_distance(e_kmer, pattern) <= d:
                pattern1.append(e_kmer)

        for pat in pattern1:
            if check_seq(text,pat,d) == True:
                all_motifs.append(pat)
#         break   
    
    all_motifs = list(set(all_motifs))
    all_motifs.sort()
    print(all_motifs)                       
    return all_motifs
        
def main():
    with open('/Users/nises/Downloads/dataset_604416_8.txt') as input_data:
            text1 = input_data.read().splitlines()
    k = int(text1[0].split(' ')[0])
    d = int(text1[0].split(' ')[1])
    text = text1[1]
    text = text.split(' ')
    
#     print(text)
#     print(k)
#     print(d)
        
#     text = ['ATTTGGC', 'TGCCTTA', 'CGGTATC', 'GAAAATT']
#     print(text)
    answer = motif_gen(text,k,d)
    
    

    with open(os.path.join('/Users/nises/Desktop/Answer.txt'), 'w') as output_data:
        output_data.write(' '.join(answer))


if __name__ == '__main__':
    main()
```

```python
import os
import collections

def min_score(pattern, motif):
    m_score = [] 
    
    for i in range(len(motif)-len(pattern)+1):
        m_score.append(calculate_ham_distance(motif[i:i+len(pattern)], pattern))
                
    return min(m_score)

def median_string(text, k):
    best_pattern = [k*len(text) + 1]
    
    possible_kmers = all_combos(k)
    
    for e_kmer in possible_kmers:
        current_score = sum([min_score(''.join(e_kmer),dna) for dna in text])
        if current_score < best_pattern[0]:
            best_pattern = [current_score, ''.join(e_kmer)]
                         
    return best_pattern[1]
        
def main():
    with open('/Users/nises/Downloads/dataset_604418_9.txt') as input_data:
            text1 = input_data.read().splitlines()
    k = int(text1[0])
    text = text1[1:]
    
#     print(text)
#     print(k)
    
    answer = median_string(text,k)
    print(answer)
    
    with open(os.path.join('/Users/nises/Desktop/Answer.txt'), 'w') as output_data:
        output_data.write(''.join(answer))

if __name__ == '__main__':
    main()
```
