```python
import functools
def suffix_array(text):
    l = len(text)
    def sort_i(i,j):        
        while i<l and j<l:
            if text[i]>text[j]:
                return 1
            elif text[i]<text[j]:
                return -1
            i +=1
            j +=1
        return 0
    return sorted(range(len(text)), key = functools.cmp_to_key(sort_i))

suffix_array('AACGATAGCGGTAGA$')
```
